```python
import pandas as pd
import numpy as np
import warnings

warnings.filterwarnings("ignore", category=FutureWarning)

raw_data = """1,Mickéy Mousé,56,70kgs,72,69,71,-,-,-\n2,Donald Duck,34,154lbs,68,69,70,-,-,-\n5,Bryan,52,180lbs,69,70,72,-,-,-\n,Plastic Man,,61kgs,-,-,-,-,-,-\n3,Super Man,19,160lbs,-,-,-,69,70,71\n2,Donald Duck,34,154lbs,68,69,70,-,-,-\n4,Batman,21,132lbs,68,69,70,72,75,76\n"""

with open("patient_heart_rate.csv", "w", encoding="utf-8") as f:
    f.write(raw_data)

print("Đã khởi tạo file 'patient_heart_rate.csv' thành công!\n")


column_names = ["Id", "Name", "Age", "Weight", "m0006", "m0612", "m1218", "f0006", "f0612", "f1218"]

df = pd.read_csv("patient_heart_rate.csv", names=column_names)

print("--- Kết quả Vấn đề 1: Bổ sung Header ---")
print(df.head())
print("-" * 50)


df[['Firstname', 'Lastname']] = df['Name'].str.split(expand=True)

df = df.drop('Name', axis=1)

print("--- Kết quả Vấn đề 2: Tách cột Name thành Firstname và Lastname ---")
print(df.head())
print("-" * 50)


for i in range(0, len(df)):
    x = str(df.at[i, 'Weight'])

    if "lbs" in x[-3:]:
        x = x[:-3:]
        float_x = float(x)

        y = int(float_x / 2.2)
        y = str(y) + "kgs"
        df.at[i, 'Weight'] = y

print("--- Kết quả Vấn đề 3: Chuẩn hóa đơn vị đo cột Weight về kgs ---")
print(df.head())
print("-" * 50)


df.dropna(how="all", inplace=True)

print("--- Kết quả Vấn đề 4: Xóa bỏ các dòng trống hoàn toàn ---")
print(df.head())
print("-" * 50)


df = df.drop_duplicates(subset=['Firstname', 'Lastname', 'Age', 'Weight'])

print("--- Kết quả Vấn đề 5: Loại bỏ các dòng trùng lặp dữ liệu ---")
print(df)
print("-" * 50)


def remove_non_ascii(text):
    if isinstance(text, str):
        return text.encode('ascii', 'ignore').decode('ascii')
    return text

df['Firstname'] = df['Firstname'].apply(remove_non_ascii)
df['Lastname'] = df['Lastname'].apply(remove_non_ascii)

print("--- Kết quả Vấn đề 6: Loại bỏ ký tự đặc biệt lỗi font ---")
print(df)
print("-" * 50)


df = pd.melt(df, id_vars=["Id", "Age", "Weight", "Firstname", "Lastname"],
             value_name="PulseRate", var_name="sex_and_time").sort_values(['Id', 'Age', 'Weight', 'Firstname', 'Lastname'])

print("--- Kết quả Vấn đề 7: Biến đổi cấu trúc dữ liệu với hàm Melt ---")
print(df.head(10))
print("-" * 50)


tmp_df = df["sex_and_time"].str.extract(r"(\D)(\d+)(\d{2})", expand=True)
tmp_df.columns = ["sex", "hours_lower", "hours_upper"]

tmp_df["Time"] = tmp_df["hours_lower"] + "-" + tmp_df["hours_upper"]

df = pd.concat([df, tmp_df], axis=1)

df = df.drop(["sex_and_time", "hours_lower", "hours_upper"], axis=1)

print("--- Kết quả Vấn đề 8: Trích xuất các thuộc tính Giới tính và Thời gian ---")
print(df.head(10))
print("-" * 50)


df.replace('-', np.nan, inplace=True)

df = df.dropna(subset=['PulseRate'])

df = df[['Id', 'Firstname', 'Lastname', 'Age', 'Weight', 'sex', 'Time', 'PulseRate']]

df.to_csv("outputcleanup.csv", index=False)

print("--- KẾT QUẢ CUỐI CÙNG (DỮ LIỆU ĐÃ LÀM SẠCH HOÀN TOÀN) ---")
print(df)
print("\n Chúc mừng bạn! Đã xuất file dữ liệu sạch thành công: 'outputcleanup.csv'")
```
