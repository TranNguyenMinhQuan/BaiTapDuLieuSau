```python
from nltk.corpus import stopwords

english_stops = stopwords.words('english')
print("Câu 1: Danh sách 20 stopword tiếng Anh đầu tiên trong NLTK:")
print(f"   {english_stops[:20]}\n")
```

```python
print(f"Câu 2: Tổng số lượng stopword tiếng Anh trong NLTK là: {len(english_stops)} từ.\n")
```

```python
print("Câu 3: Số lượng từ dừng ở một số ngôn ngữ khác:")
for lang in ['french', 'german', 'spanish', 'italian']:
    print(f"   - Ngôn ngữ {lang.capitalize()}: {len(stopwords.words(lang))} từ dừng.")
print()
```

```python
import nltk
nltk.download('punkt_tab')
text_sample = "This is a simple sample sentence showing how to filter out stop words."
words_sample = nltk.word_tokenize(text_sample)
filtered_sample = [w for w in words_sample if w.lower() not in english_stops]
print("Câu 4:")
print(f"   + Văn bản gốc: {text_sample}")
print(f"   + Sau khi lọc bỏ stopword: {filtered_sample}\n")
```

```python
words_to_keep = {'not', 'no'}
custom_stops = [w for w in stopwords.words('english') if w not in words_to_keep]
print(f"Câu 5: Số lượng stopword sau khi đã loại trừ 'not' và 'no': {len(custom_stops)}")
```

```python
from nltk.corpus import wordnet as wn

word_lookup = "travel"
synsets = wn.synsets(word_lookup)
print(f"Câu 6: Tra cứu từ '{word_lookup}' trong WordNet:")
if synsets:
    print(f"   + Định nghĩa: {synsets[0].definition()}")
    if synsets[0].examples():
        print(f"   + Ví dụ sử dụng: {synsets[0].examples()[0]}")
print()
```

```python
synonyms = set()
antonyms = set()
for syn in wn.synsets("good"):
    for l in syn.lemmas():
        synonyms.add(l.name())
        if l.antonyms():
            antonyms.add(l.antonyms()[0].name())
print("Câu 7: Từ 'good'")
print(f"   + Các từ đồng nghĩa: {list(synonyms)[:5]}...")
print(f"   + Các từ trái nghĩa: {list(antonyms)[:5]}...\n")
```

```python
print("Câu 8: Chi tiết về nhãn từ loại 'NN' (Noun - Danh từ):")
nltk.help.upenn_tagset('NN')
print()
```

```python
noun1 = wn.synset('dog.n.01')
noun2 = wn.synset('cat.n.01')
noun_sim = noun1.path_similarity(noun2)
print(f"Câu 9: Độ tương đồng ngữ nghĩa giữa danh từ 'dog' và 'cat' là: {noun_sim:.4f}\n")
```

```python
verb1 = wn.synset('run.v.01')
verb2 = wn.synset('walk.v.01')
verb_sim = verb1.path_similarity(verb2)
print(f"Câu 10: Độ tương đồng ngữ nghĩa giữa động từ 'run' và 'walk' là: {verb_sim:.4f}")
```

```python
from nltk.corpus import names

male_names = names.words('male.txt')
female_names = names.words('female.txt')
print("Câu 11:")
print(f"   + Tổng số lượng tên Nam (male.txt): {len(male_names)} tên.")
print(f"   + Tổng số lượng tên Nữ (female.txt): {len(female_names)} tên.\n")
```

```python
print("Câu 12:")
print(f"   + 10 tên Nam đầu tiên: {male_names[:10]}")
print(f"   + 10 tên Nữ đầu tiên: {female_names[:10]}\n")
```

```python
macbeth_words = nltk.corpus.gutenberg.words('shakespeare-macbeth.txt')
macbeth_text = nltk.Text(macbeth_words)

print("Câu 13: 15 cụm kết hợp từ phổ biến nhất (Collocations) trong tác phẩm Macbeth:")
macbeth_text.collocations(num=15)
```
