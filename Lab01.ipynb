```python
import pandas as pd
import numpy as np

df = pd.read_csv('dulieuxettuyendaihoc.csv')

print("10 DÒNG ĐẦU TIÊN")
display(df.head(10))

print("10 DÒNG CUỐI CÙNG")
display(df.tail(10))
```

```python
print("Bảng tần số cột DT (kể cả NaN):")
print(df['DT'].value_counts(dropna=False))

print("\nBảng tần suất cột DT (kể cả NaN):")
print(df['DT'].value_counts(dropna=False, normalize=True))

print("\nCác giá trị riêng biệt trong cột DT:")
print(df['DT'].unique())

df['DT'] = df['DT'].fillna(0)

print("\nSố lượng giá trị thiếu cột DT sau khi xử lý:", df['DT'].isnull().sum())
```

```python
print("Số lượng giá trị thiếu cột T1 trước xử lý:", df['T1'].isnull().sum())

t1_mean = df['T1'].mean()
df['T1'] = df['T1'].fillna(t1_mean)

print("Số lượng giá trị thiếu cột T1 sau khi xử lý:", df['T1'].isnull().sum())
```

```python
diem_cols = [
    'T1', 'L1', 'H1', 'S1', 'V1', 'X1', 'D1', 'N1',
    'T2', 'L2', 'H2', 'S2', 'V2', 'X2', 'D2', 'N2',
    'T6', 'L6', 'H6', 'S6', 'V6', 'X6', 'D6', 'N6',
    'DH1', 'DH2', 'DH3'
]

for col in diem_cols:
    if col in df.columns:
        mean_val = df[col].mean()
        df[col] = df[col].fillna(mean_val)

print("Kiểm tra lại tổng số giá trị thiếu của các cột điểm số:")
print(df[diem_cols].isnull().sum().sum())
```

```python
df['TBM1'] = (df['T1']*2 + df['L1'] + df['H1'] + df['S1'] + df['V1']*2 + df['X1'] + df['D1'] + df['N1']) / 10
df['TBM2'] = (df['T2']*2 + df['L2'] + df['H2'] + df['S2'] + df['V2']*2 + df['X2'] + df['D2'] + df['N2']) / 10
df['TBM3'] = (df['T6']*2 + df['L6'] + df['H6'] + df['S6'] + df['V6']*2 + df['X6'] + df['D6'] + df['N6']) / 10

df[['TBM1', 'TBM2', 'TBM3']].head()
```

```python
dieu_kien_xep_loai = [0, 5.0, 6.5, 8.0, 9.0, 10.0]
nhan_xep_loai = ['Y', 'TB', 'K', 'G', 'XS']

df['XL1'] = pd.cut(df['TBM1'], bins=dieu_kien_xep_loai, labels=nhan_xep_loai, right=False)
df['XL2'] = pd.cut(df['TBM2'], bins=dieu_kien_xep_loai, labels=nhan_xep_loai, right=False)
df['XL3'] = pd.cut(df['TBM3'], bins=dieu_kien_xep_loai, labels=nhan_xep_loai, right=False)

df[['TBM1', 'XL1', 'TBM2', 'XL2', 'TBM3', 'XL3']].head()
```

```python
df['US_TBM1'] = ((df['TBM1'] - 0) / (10 - 0)) * 4
df['US_TBM2'] = ((df['TBM2'] - 0) / (10 - 0)) * 4
df['US_TBM3'] = ((df['TBM3'] - 0) / (10 - 0)) * 4

df[['TBM1', 'US_TBM1', 'TBM2', 'US_TBM2', 'TBM3', 'US_TBM3']].head()
```

```python
def tinh_kqxt(row):
    kt = str(row['KT']).strip().upper()
    dh1, dh2, dh3 = row['DH1'], row['DH2'], row['DH3']

    if kt in ['A', 'A1']:
        diem_xt = (dh1 * 2 + dh2 + dh3) / 4
    elif kt == 'B':
        diem_xt = (dh1 + dh2 * 2 + dh3) / 4
    else:
        diem_xt = (dh1 + dh2 + dh3) / 3

    return 1 if diem_xt >= 5.0 else 0

df['KQXT'] = df.apply(tinh_kqxt, axis=1)

print(df['KQXT'].value_counts())
df[['KT', 'DH1', 'DH2', 'DH3', 'KQXT']].head(10)
```

```python
df.to_csv('processed_dulieuxettuyendaihoc.csv', index=False)
print("Đã lưu file processed_dulieuxettuyendaihoc.csv thành công!")
```
